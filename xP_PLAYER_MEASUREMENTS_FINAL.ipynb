```python
import pandas as pd
import numpy as np
```

```python
#this dataframe has the xPF and xPA for each player for each game
df = pd.read_excel("/Users/lb/Desktop/analytics/w_analytics/new_dtwom_22/2022_xP_player.xlsx")
#convert df["xPA"] to positive number
df["xPA"] = -df["xPA"]

#this df has the player ID, team, and number of mins played
df_players = pd.read_excel("/Users/lb/Desktop/analytics/w_analytics/new_dtwom_22/2022_players.xlsx")

#this dataframe has the xPF/40 and xPA/40 for each team
df_team = pd.read_excel("/Users/lb/Desktop/analytics/w_analytics/new_dtwom_22/2022_Team_xP_summary.xlsx")

df_team
```

```python
"""this first merges the first and second dfs loaded into this script"""

#groups by each player and sums columns
#group each players games together while adding up the xPF and xPA
df_new = df.groupby(df["pid"]).sum()
#drop Game column
df_new = df_new.drop(["Game", "xP%"], axis=1)

#Calculate new xP% metric based on total xPF and xPA
df_new["xP%"] = (df_new["xPF"]) / (df_new["xPF"] + df_new["xPA"]) *100
#df_new
merged_df = pd.merge(df_new, df_players, how="left", on="pid")
merged_df["xPF/40"] = merged_df["xPF"]/ merged_df["min"] *40
merged_df["xPA/40"] = merged_df["xPA"]/ merged_df["min"] *40
merged_df

```

```python
#merge the player dataframe with new xPF/40 and xPA/40 to their team xPF/40 and xPA/40 metrics in 'df_team'
final_df = pd.merge(merged_df, df_team, how = "left", on = "Team" )


#drop miscellaneous columns
final_df = final_df.drop(["PF_team", "PA_team", "xPF_team", "xPA_team", "P%_team"], axis=1)
#final_df

#Calculate relative xPF/40 and xPA/40
final_df["xPF/40_Rel"] = final_df["xPF/40"] - final_df["xPF/40_team"]
final_df["xPA/40_Rel"] = final_df["xPA/40"] - final_df["xPA/40_team"]
final_df["xP%_Rel"] = final_df["xP%"] - final_df["xP%_team"]
final_df
```

```python
#output df_new to folder
final_df.to_excel("/Users/lb/Desktop/analytics/w_analytics/new_dtwom_22/2022_PLAYER_xP_FINAL.xlsx")

```
